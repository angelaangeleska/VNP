# <font color = 'Orange'> Data preprocessing </font>

Context
This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

Content
The datasets consists of several medical predictor variables and one target variable, Outcome. Predictor variables includes the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.

Input columns: Pregnancies, Glucose, BloodPressure,SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age

Output columns: Outcome 0-->doesn't have diabetes / 1--> has diabetes

## <font color = 'Orange'>Detect the Missing values</font>

Count the percentage of missing values in every column of the Dataset.

```python
missing_values = df.isnull().sum()
missing_values
```

```python
percentage = missing_values / len(df) * 100
percentage
```

```python
missing_values = pd.concat([missing_values, percentage], axis=1)
missing_values.columns = ['Num. of missing values', '% of missing values']
missing_values
```

## <font color = 'Orange'>Handle the missing values</font>

## KNN

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit_transform(df['Pregnancies'].to_numpy().reshape(-1,1))
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=3)
```

```python
df['Pregnancies'] = imputer.fit_transform(df['Pregnancies'].to_numpy().reshape(-1,1))
```

```python
df['Pregnancies'].isnull().sum()
```

```python
df.isnull().sum()
```

## MICE

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
imputer = IterativeImputer(random_state=0) 
```

```python
df['BMI'] = imputer.fit_transform(df['BMI'].to_numpy().reshape(-1,1))
```

```python
df['BMI'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['SkinThickness'] = imputer.fit_transform(df['SkinThickness'].to_numpy().reshape(-1,1))
```

```python
df['SkinThickness'].isnull().sum()
```

```python
df.isnull().sum()
```

# <font color='orange'>KNN Classification</font>

## <font color = 'Orange'>Split the dataset for training and testing in ratio 80:20 </font>


```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data.iloc[:,:-1], data['Outcome'], test_size=0.2) 
```

## <font color = 'Orange'>Initialize the KNN Classification model, and use the fit function for training the model</font>

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors=5)
```

```python
classifier.fit(X_train, Y_train)
```

```python
classifier.get_params()
```

## <font color = 'Orange'>Predict the outcomes for X test</font>

```python
y_pred = classifier.predict(X_test)
y_pred
```

## <font color = 'Orange'>See the model performance, by using sklearn metrics for classification</font>



```python
from sklearn.metrics import classification_report, confusion_matrix, f1_score, accuracy_score
```

```python
print(confusion_matrix(Y_test, y_pred))
#[[True Negative (TN), False Positive (FP)],
# [False Negative (FN), True Positive (TP)]]
```

```python
print(classification_report(Y_test, y_pred))
```

```python
print(f1_score(Y_test, y_pred))
```

```python
print(accuracy_score(Y_test, y_pred))
```
